```python
## import libraries
import cvlib as cv
import cv2 
```

```python
## open webcam
mycam = cv2.VideoCapture(0) ## where 0 refers cam in lap 
```

```python
if not mycam.isOpened():
    print("Webcamp is not open")
    exit()
```

```python
## loop through frames
while mycam.isOpened():
    ## read the webcam
    status, frame = mycam.read()
    if not status:
        print("can't read the frame")
        exit()
    ## apply face detection
    face, confidence = cv.detect_face(frame) ## capturing the face and confidence
    print(face) 
    print(confidence)
    ## loop through detected faces
    for index, f in enumerate(face):
        (startx,starty) = f[0],f[1]
        (endx,endy) = f[2],f[3]
        ## draw rectangle over face
        cv2.rectangle(frame,(startx,starty), (endx,endy), (0,255,0), 2)
        text = "{:.2f}%".format(confidence[index] * 100) ## format of text in top of face rectangle
        Y = starty - 10 if starty - 10 > 10 else starty + 10 ## query - size of the text? 
        ## write confidence percentage on top of face rectange
        cv2.putText(frame,text,(startx,Y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)
    ## display output
    cv2.imshow("Real-time face detection",frame)
    ## press Q to stop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

        
    
    
```

```python
## release resources
mycam.release()
cv2.destroyAllWindows()
```
